```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

import nltk
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)

import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, classification_report,
                             roc_auc_score, roc_curve)
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import (RandomForestClassifier, AdaBoostClassifier,
                               BaggingClassifier, ExtraTreesClassifier,
                               GradientBoostingClassifier,
                               VotingClassifier, StackingClassifier)
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import shap
import pickle

print("All libraries imported successfully!")
```

## 1. Data Cleaning

```python
dataset = pd.read_csv('spam.csv', encoding='latin1')
print(f"Shape: {dataset.shape}")
dataset.head()

```

```python
# Drop irrelevant columns, rename
dataset.drop(columns=["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], inplace=True)
dataset.rename(columns={"v1": "Target", "v2": "Text"}, inplace=True)

# Label encode target  (ham=0, spam=1)
encoder = LabelEncoder()
dataset['Target'] = encoder.fit_transform(dataset['Target'])

# Remove duplicates
print(f"Duplicates before: {dataset.duplicated().sum()}")
dataset = dataset.drop_duplicates(keep="first")
print(f"Duplicates after:  {dataset.duplicated().sum()}")
print(f"Final shape: {dataset.shape}")
dataset.head()
```

## 2. Exploratory Data Analysis (EDA)

```python
# Class distribution
counts = dataset['Target'].value_counts()
print(counts)
print(f"Class imbalance ratio â Ham:{counts[0]} | Spam:{counts[1]} | Ratio: {counts[0]/counts[1]:.1f}:1")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Pie chart
axes[0].pie(counts, labels=["Ham", "Spam"], autopct="%0.1f%%",
            colors=["#4C9BE8", "#E85C5C"], startangle=140,
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[0].set_title("Class Distribution", fontsize=14, fontweight='bold')

# Bar chart
axes[1].bar(["Ham", "Spam"], counts, color=["#4C9BE8", "#E85C5C"], edgecolor='white', width=0.5)
axes[1].set_title("Message Counts", fontsize=14, fontweight='bold')
axes[1].set_ylabel("Count")
for i, v in enumerate(counts):
    axes[1].text(i, v + 30, str(v), ha='center', fontweight='bold')

plt.suptitle("â  Dataset is imbalanced (87% Ham vs 13% Spam) â will handle with SMOTE",
             fontsize=11, color='#cc0000', y=1.02)
plt.tight_layout()
plt.savefig('eda_class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Feature engineering â text statistics
dataset["Num_Character"] = dataset['Text'].apply(len)
dataset["Num_Words"]     = dataset['Text'].apply(lambda x: len(nltk.word_tokenize(x)))
dataset["Num_Sentence"]  = dataset['Text'].apply(lambda x: len(nltk.sent_tokenize(x)))

print("=== Overall Stats ===")
print(dataset[['Num_Character','Num_Words','Num_Sentence']].describe().round(2))
print("=== Ham ===")
print(dataset[dataset['Target']==0][['Num_Character','Num_Words','Num_Sentence']].describe().round(2))
print("=== Spam ===")
print(dataset[dataset['Target']==1][['Num_Character','Num_Words','Num_Sentence']].describe().round(2))
```

## 3. Text Preprocessing

```python
ps = PorterStemmer()
STOP_WORDS = set(stopwords.words('english'))

def transform_text(text):
    """
    Full NLP preprocessing pipeline:
    1. Lowercase
    2. Tokenize
    3. Remove non-alphanumeric tokens (symbols, emojis, punctuation)
    4. Remove stopwords
    5. Porter Stemming
    """
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [t for t in tokens if t.isalnum()]
    tokens = [t for t in tokens if t not in STOP_WORDS and t not in string.punctuation]
    tokens = [ps.stem(t) for t in tokens]
    return " ".join(tokens)

# Test
sample = "Congratulations! You have WON a free iPhone. Call now 08001234567 to claim your prize!!!"
print("Original:", sample)
print("Processed:", transform_text(sample))
```

```python
dataset['transformed_text'] = dataset['Text'].apply(transform_text)
print("Text preprocessing complete.")
print(f"Sample:{dataset[['Text','transformed_text','Target']].head(3).to_string()}")
```

```python
# WordCloud
try:
    from wordcloud import WordCloud

    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    for ax, label, title, cmap in zip(
            axes, [1, 0], ["SPAM", "HAM"], ["Reds", "Blues"]):
        text = dataset[dataset['Target']==label]['transformed_text'].str.cat(sep=" ")
        wc = WordCloud(width=600, height=400, background_color='white',
                       colormap=cmap, max_words=100).generate(text)
        ax.imshow(wc, interpolation='bilinear')
        ax.axis('off')
        ax.set_title(f"{title} â Most Frequent Words", fontsize=14, fontweight='bold')

    plt.tight_layout()
    plt.savefig('wordclouds.png', dpi=150, bbox_inches='tight')
    plt.show()
except ImportError:
    print("wordcloud not installed. Run: pip install wordcloud")
```

```python
# Top 30 spam vs ham words
from collections import Counter

spam_words = [w for msg in dataset[dataset['Target']==1]['transformed_text'] for w in msg.split()]
ham_words  = [w for msg in dataset[dataset['Target']==0]['transformed_text'] for w in msg.split()]

fig, axes = plt.subplots(1, 2, figsize=(18, 6))
for ax, words, title, color in zip(
        axes, [spam_words, ham_words], ['Top 30 Spam Words', 'Top 30 Ham Words'], ['#E85C5C', '#4C9BE8']):
    df_words = pd.DataFrame(Counter(words).most_common(30), columns=['word','count'])
    sns.barplot(x='word', y='count', data=df_words, ax=ax, color=color)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    ax.set_title(title, fontweight='bold')
plt.tight_layout()
plt.savefig('top_words.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Model Building

```python
tfidf = TfidfVectorizer(max_features=3000)
X = tfidf.fit_transform(dataset['transformed_text']).toarray()
y = dataset['Target'].values

print(f"Feature matrix shape: {X.shape}")
print(f"Class distribution â Ham: {(y==0).sum()} | Spam: {(y==1).sum()}")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2, stratify=y)
print(f"Train: {X_train.shape} | Test: {X_test.shape}")
```

## 5. Model Comparison â 11 Classifiers

```python
# Define all classifiers
clfs = {
    'SVC'      : SVC(kernel='sigmoid', gamma=1.0, probability=True),
    'KNN'      : KNeighborsClassifier(),
    'NaiveBayes': MultinomialNB(),
    'DecTree'  : DecisionTreeClassifier(max_depth=5),
    'LogReg'   : LogisticRegression(solver='liblinear', penalty='l1'),
    'RandomForest': RandomForestClassifier(n_estimators=50, random_state=2),
    'AdaBoost' : AdaBoostClassifier(n_estimators=50, random_state=2),
    'Bagging'  : BaggingClassifier(n_estimators=50, random_state=2),
    'ExtraTrees': ExtraTreesClassifier(n_estimators=50, random_state=2),
    'GBDT'     : GradientBoostingClassifier(n_estimators=50, random_state=2),
    'XGBoost'  : XGBClassifier(n_estimators=50, random_state=2, eval_metric='logloss', verbosity=0),
}

results = []
for name, clf in clfs.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_prob = clf.predict_proba(X_test)[:,1] if hasattr(clf, 'predict_proba') else None

    results.append({
        'Model'    : name,
        'Accuracy' : round(accuracy_score(y_test, y_pred), 4),
        'Precision': round(precision_score(y_test, y_pred), 4),
        'Recall'   : round(recall_score(y_test, y_pred), 4),
        'F1'       : round(f1_score(y_test, y_pred), 4),
        'ROC-AUC'  : round(roc_auc_score(y_test, y_prob), 4) if y_prob is not None else 'N/A',
    })

results_df = pd.DataFrame(results).sort_values('Precision', ascending=False)
print(results_df.to_string(index=False))
```

```python
# Visual comparison
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
metrics = ['Accuracy', 'Precision', 'Recall', 'F1']
colors  = ['#4C9BE8', '#E85C5C', '#2ECC71', '#F39C12']

for ax, metric, color in zip(axes.flat, metrics, colors):
    df_sorted = results_df.sort_values(metric, ascending=False)
    bars = ax.barh(df_sorted['Model'], df_sorted[metric], color=color, alpha=0.8)
    ax.set_xlabel(metric, fontweight='bold')
    ax.set_title(f"{metric} Comparison", fontweight='bold')
    ax.set_xlim(0.85, 1.01)
    for bar, val in zip(bars, df_sorted[metric]):
        ax.text(val + 0.001, bar.get_y() + bar.get_height()/2,
                f"{val:.3f}", va='center', fontsize=8)

plt.suptitle("Model Performance Comparison", fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Confusion matrix for top model (NaiveBayes â best precision)
best_clf = clfs['NaiveBayes']
y_pred_best = best_clf.predict(X_test)

cm = confusion_matrix(y_test, y_pred_best)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Ham','Spam'], yticklabels=['Ham','Spam'],
            linewidths=2, linecolor='white')
plt.title("Confusion Matrix â Multinomial Naive Bayes", fontweight='bold')
plt.ylabel("Actual"); plt.xlabel("Predicted")
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

print("Classification Report:")
print(classification_report(y_test, y_pred_best, target_names=['Ham','Spam']))
```

```python
# ROC Curves â top 5 models
plt.figure(figsize=(9, 7))
top5 = ['NaiveBayes', 'LogReg', 'SVC', 'ExtraTrees', 'XGBoost']
line_colors = ['#E85C5C','#4C9BE8','#2ECC71','#F39C12','#9B59B6']

for name, color in zip(top5, line_colors):
    clf = clfs[name]
    y_prob = clf.predict_proba(X_test)[:,1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    auc = roc_auc_score(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.3f})", color=color, linewidth=2)

plt.plot([0,1],[0,1],'k--', linewidth=1, label='Random Baseline')
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curves â Top 5 Models', fontsize=14, fontweight='bold')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('roc_curves.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Cross-Validation (Stratified 5-Fold)
> Single train-test split can be misleading. 5-fold CV gives a more reliable estimate of generalization.

```python
cv_models = {
    'NaiveBayes'  : MultinomialNB(),
    'LogReg'      : LogisticRegression(solver='liblinear', penalty='l1'),
    'ExtraTrees'  : ExtraTreesClassifier(n_estimators=50, random_state=2),
    'XGBoost'     : XGBClassifier(n_estimators=50, random_state=2, eval_metric='logloss', verbosity=0),
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = []

for name, clf in cv_models.items():
    f1_scores  = cross_val_score(clf, X, y, cv=skf, scoring='f1')
    auc_scores = cross_val_score(clf, X, y, cv=skf, scoring='roc_auc')
    cv_results.append({
        'Model'       : name,
        'CV F1 Mean'  : round(f1_scores.mean(), 4),
        'CV F1 Std'   : round(f1_scores.std(),  4),
        'CV AUC Mean' : round(auc_scores.mean(),4),
        'CV AUC Std'  : round(auc_scores.std(), 4),
    })

cv_df = pd.DataFrame(cv_results).sort_values('CV F1 Mean', ascending=False)
print(cv_df.to_string(index=False))
```

## 7. Handling Class Imbalance with SMOTE
> The dataset is ~87% Ham / 13% Spam. SMOTE (Synthetic Minority Oversampling) creates synthetic spam samples to balance training.

```python
sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)
print(f"Before SMOTE â Ham: {(y_train==0).sum()} | Spam: {(y_train==1).sum()}")
print(f"After  SMOTE â Ham: {(y_train_sm==0).sum()} | Spam: {(y_train_sm==1).sum()}")
```

```python
# Compare with/without SMOTE on NaiveBayes and LogReg
smote_results = []
for name, clf_class, kwargs in [
    ('NaiveBayes', MultinomialNB, {}),
    ('LogReg',     LogisticRegression, {'solver':'liblinear','penalty':'l1'}),
]:
    for label, Xtr, ytr in [('No SMOTE', X_train, y_train), ('SMOTE', X_train_sm, y_train_sm)]:
        clf = clf_class(**kwargs)
        clf.fit(Xtr, ytr)
        y_pred = clf.predict(X_test)
        smote_results.append({
            'Model': name, 'SMOTE': label,
            'Precision': round(precision_score(y_test, y_pred), 4),
            'Recall'   : round(recall_score(y_test, y_pred), 4),
            'F1'       : round(f1_score(y_test, y_pred), 4),
        })

smote_df = pd.DataFrame(smote_results)
print(smote_df.to_string(index=False))
print("Note: SMOTE improves Recall (catches more spam) at the cost of slight Precision drop.")
print("For spam detection, higher Recall is generally preferred.")
```

## 8. Ensemble Methods

```python
# Soft Voting Classifier
svc_ens = SVC(kernel='sigmoid', gamma=1.0, probability=True)
mnb_ens = MultinomialNB()
etc_ens = ExtraTreesClassifier(n_estimators=50, random_state=2)

voting_clf = VotingClassifier(
    estimators=[('svm', svc_ens), ('nb', mnb_ens), ('et', etc_ens)],
    voting='soft'
)
voting_clf.fit(X_train, y_train)
y_pred_v = voting_clf.predict(X_test)
print("=== Soft Voting Classifier ===")
print(f"Accuracy : {accuracy_score(y_test, y_pred_v):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_v):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred_v):.4f}")
print(f"F1       : {f1_score(y_test, y_pred_v):.4f}")
```

```python
# Stacking Classifier
stack_clf = StackingClassifier(
    estimators=[('svm', SVC(kernel='sigmoid', gamma=1.0, probability=True)),
                ('nb',  MultinomialNB()),
                ('et',  ExtraTreesClassifier(n_estimators=50, random_state=2))],
    final_estimator=LogisticRegression(),
    cv=5
)
stack_clf.fit(X_train, y_train)
y_pred_s = stack_clf.predict(X_test)
print("=== Stacking Classifier ===")
print(f"Accuracy : {accuracy_score(y_test, y_pred_s):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_s):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred_s):.4f}")
print(f"F1       : {f1_score(y_test, y_pred_s):.4f}")
```

## 9. Model Explainability â SHAP
> SHAP (SHapley Additive exPlanations) shows *why* the model made each prediction â essential for a production ML system.

```python
# SHAP on LogisticRegression (linear â fastest for SHAP)
lr_for_shap = LogisticRegression(solver='liblinear', penalty='l1', max_iter=1000)
lr_for_shap.fit(X_train, y_train)

explainer = shap.LinearExplainer(lr_for_shap, X_train, feature_perturbation='interventional')
shap_values = explainer.shap_values(X_test[:200])

feature_names = tfidf.get_feature_names_out()

plt.figure()
shap.summary_plot(shap_values, X_test[:200], feature_names=feature_names,
                  max_display=20, show=False, plot_type='bar')
plt.title("SHAP Feature Importance â Top 20 Words", fontweight='bold')
plt.tight_layout()
plt.savefig('shap_importance.png', dpi=150, bbox_inches='tight')
plt.show()
print("Red bars = spam indicators | Blue bars = ham indicators")
```

## 11. Active Learning Pipeline

```python
import os
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

print("--- INITIATING ACTIVE LEARNING PIPELINE ---")

# 1. Load the original Kaggle dataset
original_df = pd.read_csv('spam.csv', encoding="latin1")
original_df.drop(columns=["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], inplace=True, errors='ignore')
original_df.rename(columns={"v1": "Target", "v2": "Text"}, inplace=True)
# Map original labels to 0 and 1
original_df['Target'] = original_df['Target'].map({'ham': 0, 'spam': 1})
original_df = original_df[['Target', 'Text']]

# 2. Check for newly collected user feedback
if os.path.exists('retraining_data.csv'):
    print("Found new user corrections! Merging with original dataset...")
    new_df = pd.read_csv('retraining_data.csv')
    
    # Map the web UI labels ('safe' and 'spam') to 0 and 1
    new_df['Target'] = new_df['correct_label'].map({'safe': 0, 'spam': 1})
    new_df.rename(columns={'text': 'Text'}, inplace=True)
    new_df = new_df[['Target', 'Text']]
    
    # Combine the old data with the new intelligence
    combined_df = pd.concat([original_df, new_df], ignore_index=True)
    print(f"Successfully added {len(new_df)} new data points.")
else:
    print("No retraining_data.csv found. Using original dataset only.")
    combined_df = original_df

# 3. Clean the combined text (using your existing transform_text function)
print("Processing text...")
combined_df['transformed_text'] = combined_df['Text'].apply(transform_text)

# 4. Re-train the Vectorizer to learn new vocabulary
print("Retraining TF-IDF Vectorizer...")
tfidf = TfidfVectorizer(max_features=3000)
X_new = tfidf.fit_transform(combined_df['transformed_text']).toarray()
y_new = combined_df['Target'].values

# 5. Re-train the Naive Bayes Model with the new weights
print("Retraining Multinomial Naive Bayes Model...")
final_model = MultinomialNB()
final_model.fit(X_new, y_new)

# 6. Overwrite the live files
pickle.dump(tfidf, open('vectorizer.pkl', 'wb'))
pickle.dump(final_model, open('model.pkl', 'wb'))

print("â SUCCESS! model.pkl and vectorizer.pkl have been updated.")
print("Restart your FastAPI server to load the upgraded model into production.")
```
